```python
#import all the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#loading dataset
df = pd.read_csv('Indian_Traffic_Violations.csv')
print(df.shape)
print(df.head(5))
```

```python
#creating variables for violation count
location_counts = df['Location'].value_counts()
print(location_counts)
Violations_type_count = df['Violation_Type'].value_counts()
print(Violations_type_count)
Violation_vehicle_type = df['Vehicle_Type'].value_counts()
print(Violation_vehicle_type)
```

```python
#visualizing the data
plt.figure(figsize=(10,6))
location_counts.plot(kind='bar', color='pink')
plt.title("Violation by location")
plt.xlabel('Location')
plt.ylabel('Number of Violations')
plt.show()
plt.figure(figsize=(10,6))
Violations_type_count.plot(kind ='bar', color ='skyblue')
plt.title('Violation type')
plt.xlabel('Type of Violation')
plt.ylabel('Number of Violations')
plt.show()
plt.figure(figsize=(10,6))
Violation_vehicle_type.plot(kind='bar',color='purple')
plt.title('Vehicle type')
plt.xlabel('Vehicle')
plt.ylabel('Number of Violations')
plt.show()
```

```python
#creating pie chart for violation types
plt.figure(figsize=(6,6))
plt.pie(Violations_type_count, labels=Violations_type_count.index, autopct='%1.1f%%')
plt.title('Pie chart for violation types')
plt.show()
```
